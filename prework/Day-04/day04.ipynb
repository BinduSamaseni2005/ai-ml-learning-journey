
## import os

## What `os.getcwd()` Does

## `getcwd()` means:

## > **Get Current Working Directory**

## It returns the folder/path where the current Python program is running.

---

## Windows Example

## C:\\Users\\Admin
## or
## C:\Users\Admin

### About `C:\\`

### * `C:` → drive name in Windows
### * `\\` → backslash used as folder separator in Windows paths

### In Python strings, `\\` is used because `\` is an escape character.

---

## macOS / Linux Example

### /Users/admin

### or
### /home/admin


### About `/`

### * macOS and Linux use `/` as the folder separator.
### * There are no drive letters like `C:`.


## Main Difference

### | Operating System | Path Separator | Example          |
### | ---------------- | -------------- | ---------------- |
### | Windows          | `\`            | `C:\Users\Admin` |
### | macOS/Linux      | `/`            | `/Users/admin`   |


#### Escape Character in Python

#### An **escape character** is a special character used to represent characters that are difficult to type directly inside a string.

#### In Python, the escape character is:

#### (backslash)

#### It tells Python:

#### > “The next character has a special meaning.”


## Common Escape Characters

### | Escape Sequence | Meaning                |
### | --------------- | ---------------------- |
### | `\n`            | New line               |
### | `\t`            | Tab space              |
### | `\\`            | Single backslash (`\`) |
### | `\"`            | Double quote           |
### | `\'`            | Single quote           |


## Examples

### New Line

## print("Hello\nWorld")


## Output:

Hello
World

### Tab Space

### print("Python\tProgramming")

### Output:

### Python    Programming


### Backslash

### print("C:\\Users\\Admin")


### Output:

### C:\Users\Admin

### Here:

### * First `\` = escape character
### * Second `\` = actual backslash character

### So `\\` means:

### > print one backslash `\`
