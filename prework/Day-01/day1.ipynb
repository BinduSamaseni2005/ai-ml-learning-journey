#Lists


```python
simple_list =['bindu',2,3,4,2.7,'python']
print(simple_list)
print(simple_list[0])
print(simple_list[1])
print(simple_list[2])
print(simple_list[3])
print(simple_list[4])
print(simple_list[5])
print(simple_list[-1])
```

```python
#it is easy to change the values in the list
simple_list[3]=54
print(simple_list[3])
simple_list.append('to the back')
print(simple_list)
print(len(simple_list))
simple_list.pop(4)
print(simple_list)
```

```python
#it is easy to initialize a list of particular size with all the entries as the same
repeated_list =[5]*4
print(repeated_list)
```

```python
#we can even use lists as entires as list
simple_list[1] = [1,2,3]
print(simple_list)
print(simple_list[1])
print(simple_list[1][1])
```

```python
#copying lists is a little tricky
list2 = simple_list
print(list2)
print(simple_list)
list2[1]=0
print(list2)
print(simple_list)
#we see the same output because there anrent many languages dpoing this , beacuse of memory saving like c++,R
```

```python
#we must use the copy method to prevent this behaviour
list3 = simple_list.copy()
print(list3)
print(simple_list)
list3[0] = 'movi'
print(list3)
print(simple_list)
```

#Tuples

```python
#tuples are represnted by the parenthesis
#they are unchangable and allows duplicates
simple_tuple=(1,2,3,4,5)
print(simple_tuple[1])
```

```python
# it is not possible to change values in the tuple
simple_tuple[0] =5
```

```python
#this can be work around
#we can type catse tuple into list  and change the entry to the list and type cast temporary list
dummy = list(simple_tuple)
dummy[0] = 5
simple_tuple = tuple(dummy)
print(simple_tuple)
```

```python
#set is like tuple and list but it is unordered and doenot allow duplicates and it is changable
simple_set = {2,4,3,1,9}
print(simple_set)
```

```python
print(simple_set)
```

```python
print(water in simple_set)
```

```python
#you can't change values but u can add and remove the entries
simple_set.add(73)
print(simple_set)
simple_set.remove(3)
print(simple_set)
```

#dictionary

```python
# a dictionary is a collection of variables that is unordered but changable and indexed
dict = {
    "brand": "apple",
    "product": "iphone",
    "model": "X"
}
print(dict)
print('i bought an' , dict['product'],"model",dict['model'],'from',dict['brand'])
```

```python
#change things in the dictionary
dict["model"] = "11 pro"
print('i bought an',dict['product'],"model",dict['model'],'from',dict['brand'])
```

```python
#we can also add entriues to the dictionary
dict['color']='red'
print('i bought an',dict["color"],dict['product'],"model",dict['model'],'from',dict['brand'])
```
