Suppose `rand_mat` is:

rand_mat = np.array([
    [10,20,30],
    [40,50,60],
    [70,80,90]
])

Now this line:

sub_mat2 = rand_mat[0:2,0:3].copy()


means:

* rows from `0` to `2` (2 not included)
* columns from `0` to `3` (3 not included)

So NumPy selects:

[
 [10,20,30],
 [40,50,60]
]

because:

* rows selected → 0 and 1
* columns selected → 0,1,2


## Important part: `.copy()`

Without `.copy()`, NumPy creates a **view** (linked to original matrix).

With `.copy()`, NumPy creates a **separate independent matrix**.

---

# This line

sub_mat2[:] = 99


means:

> “Replace every element with 99.”

Result:


[
 [99,99,99],
 [99,99,99]
]


# So:


print(sub_mat2)


prints:


[
 [99 99 99]
 [99 99 99]
]

# What about `rand_mat`?

Because you used `.copy()`:


print(rand_mat)


still prints the original matrix unchanged:

[
 [10 20 30]
 [40 50 60]
 [70 80 90]
]


# Very important concept

## With `.copy()`

Original matrix stays safe.

## Without `.copy()`

Example:

sub_mat2 = rand_mat[0:2,0:3]
sub_mat2[:] = 99


Then `rand_mat` would ALSO change because slicing creates a view by default in NumPy.
